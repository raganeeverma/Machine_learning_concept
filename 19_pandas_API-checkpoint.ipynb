DATE RANGE FUNCTION

TIMEDELTA- FUNCTION USED TO DIFFERENCE BETWEEN DATES OR TIME

DATERANGE- RANGE BETWEEN DATES AND MONTH

```python
# DATE FORMATE for daterange MM/DD/YY

daterange=pd.date_range(start='1/1/1990', end='1/10/1990',freq='D')
daterange
```

```python

daterange=pd.date_range(start='1/1/1990', end='5/10/1990',freq='M')
daterange
```

```python
#datatime functin already present in python library

import datetime

dt=datetime.datetime.now()#present date
print(dt)

#date after 10 days

#time delta present in pandas used to the defferent between two dates and time

print(dt+pd.to_timedelta(10,unit='D'))
```
