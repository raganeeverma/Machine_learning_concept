handle missing value

2 Delete Rows with Missing Values:

3. other method fillna,inplace


```python
from sklearn.impute import SimpleImputer

imp = SimpleImputer(strategy='most_frequent')#nan value replaced by mode value

df['City']=imp.fit_transform(df['City'].values.reshape(-1,1))

df
```

```python
imp = SimpleImputer(strategy='mean')#nan value replaced by mode value

df['Age']=imp.fit_transform(df['Age'].values.reshape(-1,1))

df
```

```python
# replacing missing values in salary
# column with mean of that column
df['Salary'] =df['Salary'].fillna(df['Salary'].mean())
```
