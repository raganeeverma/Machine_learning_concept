```python
from scipy.stats import zscore
```

```python
z=np.abs(zscore(dscar_new))  #abs is absolute function used to make all data positive 

#in z score value is both value -3 and +3 is present but abs function convert -3 value to positive for better calculation
z
```

```python
threshold=3
print(np.where(z>3))
```

```python
dscar_new1=dscar[(z<3).all(axis=1)]#all rows for each column
dscar_new1 #these are normal data or clean data
```
